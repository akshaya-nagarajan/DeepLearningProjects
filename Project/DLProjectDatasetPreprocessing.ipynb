## Change all the File Names - according to the folder name, for identifying clearly.

```
def changeFileNames(fileList, directoryName):
  DRYRUN = False
  for (index,filename) in enumerate(fileList):
    extension = os.path.splitext(filename)[1]
    newname = "%s_%d%s" % (directoryName, index, extension)    
    if os.path.exists(newname):
      print("Cannot rename %s to %s, already exists" % (filename,newname))
      continue
    if DRYRUN:
      print("Would rename %s to %s" % (filename,newname))
    else:
      print("Renaming %s to %s" % (filename,newname))
      print(filename)
      os.rename(filename,newname)
```

## Parse each image, get only the images with humans in them. Delete all other Images from the Dataset. As the Dataset was directly collected by downlading from Google, it'll contain many images unrelated to Human Yoga Poses. Also, the images are cropped, so that it'll contain only one human image in it. Apart from these operations, the file formats are changed to JPG, to maintain uniformity.

```
def detect_persons_and_save_crops(fileName, graph_def, folder, newFolder, newFileName):
    with tf.Session() as sess:
        # Restore session
        sess.graph.as_default()
        tf.import_graph_def(graph_def, name='')

        files = list(Path(folder).glob('*.png'))
        files.extend(list(Path(folder).glob('*.jpg')))

        for fid, fname in enumerate(files): 
            # Read and preprocess an image.
            img = cv.imread(str(fname))
            if(img is not None):
                rows = img.shape[0]
                cols = img.shape[1]
                inp = cv.resize(img, (300, 300))
                inp = inp[:, :, [2, 1, 0]]  # BGR2RGB

                # Run the model
                out = sess.run([sess.graph.get_tensor_by_name('num_detections:0'),
                                sess.graph.get_tensor_by_name('detection_scores:0'),
                                sess.graph.get_tensor_by_name('detection_boxes:0'),
                                sess.graph.get_tensor_by_name('detection_classes:0')],
                               feed_dict={'image_tensor:0': inp.reshape(1, inp.shape[0], inp.shape[1], 3)})

                # Visualize detected bounding boxes.
                num_detections = int(out[0][0])
                #print(fname, num_detections)
                #newFolder = '/content/gdrive/My Drive/Deep Learning project/Images/WarriorPoseCleaned'
                for i in range(num_detections):
                    classId = int(out[3][0][i])
                    if(classId==1):
                        score = float(out[1][0][i])
                        bbox = [float(v) for v in out[2][0][i]]
                        if score > 0.3:
                            x = int(bbox[1] * cols)
                            y = int(bbox[0] * rows)
                            right = int(bbox[3] * cols)
                            bottom = int(bbox[2] * rows)
                            writeFile = str(newFolder) + str(newFileName)+ str(fid) + '-' + str(i) +'.jpg'
                            print(writeFile)
                            cv.imwrite(writeFile, img[y:bottom, x:right])
                           

            try:
                send2trash(str(fname))
            except: pass    
        print('Cleaning done.')
    return
```
